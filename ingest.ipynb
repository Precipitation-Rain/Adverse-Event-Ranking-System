## Data cleaning steps

### Handling Duplicates

```python
df.duplicated(subset=['primaryid','drugname', 'reaction']).sum()
```

```python
df.drop_duplicates(subset=['primaryid','drugname', 'reaction'] , inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['drugname'] =  df['drugname'].str.lower()
df['reaction'] = df['reaction'].str.lower()
```

```python
df
```

### Handling missing values and outcome_code columns

```python
df['outcome_code'] = df['outcome_code'].fillna('UN')
```

```python
df['outcome_code'] = df['outcome_code'].str.strip()
```

```python
df.isnull().sum()
```

```python
df['outcome_code'].value_counts()
```

```python
dictionary = {
    'DE' : 10,
    'LT' : 8,
    'HO' : 7,
    'DS' : 6,
    'CA' : 5,
    'RI' : 4,
    'OT' : 2,
    'UN' : 2
}
```

```python
df['severity_score'] = df['outcome_code'].map(dictionary)
```

```python
df
```

```python
df['outcome_code'].value_counts()
```

```python
df['severity_score'].value_counts()
```

```python
df
```

```python
df['country'] = df['country'].fillna('NN')
```

```python
df.isnull().sum()
```

### Handling date columns

```python
df['report_date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df
```

```python
df.drop(columns={'date' , 'report_date'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```
